```python
from yelpapi import YelpAPI
import json
import pandas as pd
```

```python
with open('/Users/caell/.secret/yelp_api.json', 'r') as f:
    login = json.load(f)
login.keys()
```

```python
yelp_api = YelpAPI(login['api-key'], timeout_s=5)
```

```python
results = yelp_api.search_query(location='San Diego, CA',
                               term='mexican food')
results.keys()
```

```python
results['total']
```

```python
mexican1 = pd.DataFrame(results['businesses'])
mexican1.head()
```

```python
results2 = yelp_api.search_query(location='San Diego, CA',
                                term='mexican food',
                                offset=20)
results2.keys()
```

```python
mexican2 = pd.DataFrame(results2['businesses'])
mexican2.head()
```
