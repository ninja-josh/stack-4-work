```python
# Imports
import pandas as pd
import json
```

```python
# Load and examine the info file
super_info = pd.read_csv('Data/superhero_info - superhero_info.csv')
super_info.info()
super_info.head()
```

# Transform Superhero Info

## Split Hero and Publisher

```python
# split the Hero/Publisher column into 2 columns and drop 'Hero|Publisher'
super_info[['Hero', 'Publisher']] = \
super_info['Hero|Publisher'].str.split('|', expand=True)
super_info.drop('Hero|Publisher', axis=1, inplace=True)

super_info.head()
```

## Split Measurements

```python
# Check datatype of 'Measurements'
super_info.loc[0, 'Measurements']
```

```python
# Replace single quotes with double quotes for json
super_info['Measurements'] = super_info['Measurements'].str.replace("'", '"')
super_info.loc[0, 'Measurements']
```

```python
# Transform the string to a list
super_info['Measurements'] = super_info['Measurements'].apply(json.loads)
```

```python
# Expand the list to separate columns, Height and Weight
super_info[['Height', 'Weight']] = super_info['Measurements'].apply(pd.Series)
super_info.drop('Measurements', axis=1, inplace=True)
super_info.head()
```

## Convert Height and Weight to numeric

```python
# Clean up Height and Weight and convert to float
super_info['Height'] = super_info['Height'].str.replace(' cm', '').astype(float)
super_info['Weight'] = super_info['Weight'].str.replace(' kg', '').astype(float)
super_info.head()
```

# Super_powers

```python
# Load and examine the powers file
super_powers = pd.read_csv('Data/superhero_powers - superhero_powers.csv')
super_powers.info()
super_powers.head()
```

```python
# Examine the first element in Powers
super_powers.loc[0,'Powers']
```

```python
# Convert the string of powers to a list of powers
Powers = super_powers['Powers'].str.split(',')
```

```python
# explode the list of powers into one row per hero per power
exploded = Powers.explode('Powers')

# Get a list of the unique powers
cols_to_make = exploded.dropna().unique()
cols_to_make
```

```python
# One-hot encode the powers
for col in cols_to_make:
    super_powers[col] = super_powers['Powers'].str.contains(col)
super_powers
```

```python
# defragging the dataframe
super_powers = super_powers.copy()
```

```python
# merge info and powers dataframes
SuperHeroes = pd.merge(super_info, super_powers, 
                       left_on='Hero', right_on='hero_names')
SuperHeroes.head()
```

```python
SuperHeroes.info()
```

# Questions:

## Compare the average weight of super powers who have Super Speed to those who do not.

```python
# replacing True and False with 1 and 0
# Not sure if this is necessary, but it's what I'm used to seeing
SuperHeroes = SuperHeroes.replace([True, False], [1, 0])
```

```python
# create a df of just speedy heroes
speed = SuperHeroes[SuperHeroes['Super Speed'] == True]
speed['Super Speed'].value_counts()
```

```python
# create a df of slo
no_speed = SuperHeroes[~(SuperHeroes['Super Speed'] == True)]
no_speed.head()
```

```python
print('Speedy Weights = ', speed['Weight'].mean())
print('Slow Weights = ', no_speed['Weight'].mean())
```
